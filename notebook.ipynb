```python
stock = {
    "pains au chocolat": 500,
    "croissants": 400,
    "baguettes": 500,
    "beignets": 200
}
```

```python
# Nouveau code propre pour remplacer le précédent
def visualiser_inventaire(inventaire):
    print("--- Rapport d'Inventaire ---")
    for produit, quantite in inventaire.items():
        # Utilisation de capitalize() pour un affichage propre
        print(f"Produit : {produit.capitalize()} | Quantité : {quantite}")

# Test avec votre dictionnaire
visualiser_inventaire(stock)
```
