```python
stock = {
    "pains au chocolat": 500,
    "croissants": 400,
    "baguettes": 500,
    "beignets": 200
}
```

```python
def afficher_inventaire(donnees):
    for produit, quantite in donnees.items():
        print(f"{produit.capitalize()} : {quantite}")

# Test de la fonction
afficher_inventaire(stock)
```
